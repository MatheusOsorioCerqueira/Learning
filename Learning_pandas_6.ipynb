```python
import numpy as np
import pandas as pd
```

```python
ind = list(np.arange(0,12))
d1 = {
    'A':'A1 A2 A3 A4'.split(),
    'B':'B1 B2 B3 B4'.split(),
    'C':'C1 C2 C3 C4'.split(),
    'D':'D1 D2 D3 D4'.split()   
}
```

```python
df1 = pd.DataFrame(d1, index = ind[:4])
df1
```

```python
df2 = pd.DataFrame(d1, index = ind[4:8])
df2
```

```python
df3 = pd.DataFrame(d1,index[8:])
df3
```

```python
#para concatenar dataframes temos o metodo concat
pd.concat([df1,df3])
```

```python
#para fazer uma junção pelas colunas usamos o axis=1
pd.concat([df1,df3],axis=1)#o motivo plo qual ele nao virou um dataframe 4x8 é porque os índices sao diferentes
```

```python
#para "juntar" as colunas temos o metodo merge
#se nao tivessemos o on='A' ele juntaria todas as colunas iguais
#nesse metodo os DataFrames sao passados como parametros
pd.merge(df1,df2,on='A')
```

```python
my_dict = {
    'E': 'E1 E2 E3 E4'.split(),
    'F': 'F1 F2 F3 F4'.split(),
    'G': 'G1 G2 G3 G4'.split()
}
df3 = pd.DataFrame(my_dict,index = [0,1,3,5])
```

```python
#o metodo join da um concat apenas se os valores do indice forem iguais
df1.join(df3)
```

```python
#se usarmos o metodo "outer" veremos que ambos os indices são levados em conta
df1.join(df3,how='outer')
```

```python
import matplotlib.pyplot as plt #padrao de importação do pyplot
import numpy as np
```

```python
%matplotlib inline #metodo relevante apenas para o jupyter notebook
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = np.array(list(map(lambda x: x*x + 2,x)))
y
```

```python
plt.plot(x,y,'r--')#faz o grafico,o parametro color é opcional
plt.xlabel("Eixo X")#define o que será mostrado no eixo X
plt.ylabel("Eixo Y")#define o que será mostrado no eixo Y
plt.show()
```

```python
plt.subplot(1,2,1)#cria mais de um plot, o ultimo argumento determina com qual plot você estará trabalhando
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
fig = plt.figure()#faz um objeto do tipo figura
axes1 = fig.add_axes([0.5,0.5,0.5,0.5])#adiciona uma figura na posição determinada. os valores representam [pos_x,pos_y,tam_x,tam_y]
axes1.set_xlabel("Eixo X")#define o label de x
axes1.set_title("Titulo")#define o titulo do axe
axes2 = fig.add_axes([0.6,0.8,0.1,0.1])
axes2.plot(y,x)
axes1.plot(x,y)
```
